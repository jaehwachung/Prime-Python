**반지름이 0보다 클 때까지 계속**


```python
# 원의 넓이와 둘레 계산 프로그램
#반지름 사용자 입력
rad = int(input("반지름을 입력하세요: "))
if rad > 0 :
      #넓이 출력
      print(f"원의 넓이는 {rad ** 2 * 3.141:.1f} 입니다.")
      #둘레 출력
      print(f"원의 둘레는 {2 * rad * 3.141:.1f} 입니다.")
else :
      # rad가 0보다 작을때
      print("0보다 큰 수를 입력하세요.")
```
