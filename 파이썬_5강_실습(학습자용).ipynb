```python
# 원의 넓이와 둘레 계산 프로그램
rad = int(input("반지름을 입력하세요: "))

#넓이 출력
print(f"원의 넓이는 {rad ** 2 * 3.141:.2f} 입니다.")
#둘레 출력
print(f"원의 둘레는 {2 * rad * 3.141:.2f} 입니다.")
```
